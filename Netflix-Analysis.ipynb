# 6.WHAT IS THE MONTHLY TREND OF NETFLIX CONTENT ADDED OVER THE YEARS?

```python
nt['month_added'] = nt['date_added'].dt.month_name()
monthy_trend = nt.groupby(['year_added', 'month_added']).size().reset_index(name='count')

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

monthy_trend['month_added'] = pd.Categorical(monthy_trend['month_added'], categories=months, ordered=True)

monthy_trend = monthy_trend.sort_values(['year_added', 'month_added'])

years = monthy_trend['year_added'].unique()

plt.figure(figsize=(12, 6))
for i in years:
    data = monthy_trend[monthy_trend['year_added'] == i]
    plt.plot(data['month_added'], data['count'], marker="*", label=str(i))

plt.title('Monthly Trend in Each Year')
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

# 8.WHAT IS THE AVERAGE DURATION OF MOVIES BY COUNTRY?

```python
avg_dur=nt[nt['type']=='Movie']
avg_dur['duration']=avg_dur['duration'].str.extract(r'(\d+)')
avg_dur = avg_dur.dropna(subset=['country', 'duration'])
avg_dur['country'] = avg_dur['country'].str.split(', ')
avg_dur = avg_dur.explode('country')
avg_dur['duration']=avg_dur['duration'].astype('float')
avg_dur1=avg_dur.groupby('country')['duration'].mean().reset_index()
avg_dur1 = avg_dur1.sort_values(by='duration', ascending=False)

avg_dur1[:10].plot(kind='bar',x='country',y='duration',color='turquoise')
plt.xlabel('duration')
plt.title('average duration of movies')
plt.show()
```

# 9. HOW MANY TITLES ON NETFLIX HAVE “LOVE” IN THEIR TITLE? SHOW A YEARLY TREND.

```python
love_title=nt[nt['title'].str.contains('love',case=False)]
total=love_title.groupby('year_added')['title'].size().reset_index()
total.plot(kind='bar',x='year_added',y='title',color='turquoise')
plt.ylabel('title count')
plt.legend()
plt.show()
```

# 10. WHAT IS THE DISTRIBUTION OF CONTENT TYPES (MOVIES VS TV SHOWS) ACROSS DIFFERENT COUNTRIES?

```python
tvs_mov=nt.dropna(subset=['country','type'])
tvs_mov1=tvs_mov.groupby(['country','type']).size().unstack(fill_value=0)
tvs_mov2=tvs_mov1.sum(axis=1).sort_values(ascending=False).head(10)
tvs_mov1.loc[tvs_mov2.index].plot(kind='bar',figsize=(12,6))
plt.title('Movies vs TV Shows in Top 10 Content-Producing Countries')
plt.ylabel('Number of Titles')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

# 12.WHAT IS THE DISTRIBUTION OF MOVIES BY DURATION CATEGORIES?

```python
movies=nt[nt['type']=='Movie'].copy()
movies['duration_mins'] = movies['duration'].str.extract(r'(\d+)').astype(float)
bins=[0,60,90,120,float('inf')]
labels=['<60 min','60-90 min','90-120 min','120 + min']
movies['duration_category'] = pd.cut(movies['duration_mins'], bins=bins, labels=labels)
movies['duration_category'].value_counts().sort_index().plot(kind='bar', color='turquoise')
plt.title('Distribution of Movies by Duration')
plt.xlabel('Duration Category')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

# 13.WHAT ARE THE MOST COMMON GENRES FOR MOVIES AND TV SHOWS?

```python
defcon=nt.dropna(subset=['listed_in'])
defcon['listed_in']=defcon['listed_in'].str.split(',')
defcon=defcon.explode('listed_in')
defcon1=defcon.groupby(['type','listed_in']).size().reset_index(name='count')
mov_gen=defcon1[defcon1['type']=='Movie'].sort_values('count',ascending=False).head(10)
sho_gen=defcon1[defcon1['type']=='TV Show'].sort_values('count',ascending=False).head(10)

plt.figure(figsize=(10,6))


plt.subplot(1,2,1)
plt.bar(mov_gen['listed_in'],mov_gen['count'],color='teal')
plt.xlabel('movie genre')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(10,6))
plt.subplot(1,2,2)
plt.bar(sho_gen['listed_in'],sho_gen['count'],color='turquoise')
plt.xlabel('TV Show genre')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.show()
```

# 14.HOW DOES THE AVERAGE DURATION OF MOVIES VARY ACROSS THE TOP 10 MOST COMMON GENRES?

```python
movie_duration=nt.dropna(subset=['duration','listed_in'])
movie_duration=movie_duration[movie_duration['type']=='Movie'].copy()
movie_duration['duration']=movie_duration['duration'].str.extract(r'(\d+)').astype('float')
movie_duration['listed_in']=movie_duration['listed_in'].str.split(',')
movie_duration=movie_duration.explode('listed_in')
movie_duration['listed_in']=movie_duration['listed_in'].str.strip()
avg_duration=movie_duration.groupby('listed_in')['duration'].mean().sort_values(ascending=False).head(10)
avg_duration.reset_index().plot(kind='bar',x='listed_in',y='duration',color='teal')
plt.ylabel('duration')
plt.title('average duration for each genre')
plt.xticks(rotation=90)
plt.show()

```

# 15.WHICH ARE THE TOP 10 WORDS THAT MOST FREQUENTLY APPEAR IN THE TITLES OF NETFLIX CONTENT?

```python
most_title=nt.dropna(subset=['title'])
most_title['title']=nt['title'].str.lower()
most_title['title']=most_title['title'].str.split(' ')
most_title_ex=most_title.explode('title')
word_counts=most_title_ex['title'].value_counts().reset_index()
word_counts.columns=['word','count']
import string
stopwords = {
    'the', 'of', 'a', 'in', 'and', 'to', 'my', 'on', 'for', 'by', 'with',
    'from', 'at', 'is', 'it', 'an', 'as', '&'
}
most_title_ex = most_title_ex[~most_title_ex['title'].str.isnumeric()]
most_title_ex['title'] = most_title_ex['title'].str.strip(string.punctuation)
most_title_ex = most_title_ex[most_title_ex['title'].str.strip() != '']
most_title_ex = most_title_ex[~most_title_ex['title'].isin(stopwords)]
cleaned_word_counts = most_title_ex['title'].value_counts().reset_index()
cleaned_word_counts.columns = ['word', 'count']
cleaned_word_counts.head(10).plot(kind='bar',x='word',y='count',color='turquoise')
plt.ylabel('count')
plt.show()
```
